### Amazon En Çok Satanlar Listesi İçeriğini Alma

```python
import requests
from bs4 import BeautifulSoup
```

```python
sayfaURL="https://www.amazon.com.tr/gp/bestsellers/books/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for kitapLi in soup.find_all("li",{"class":"zg-item-immersion"}):
        kitapIsmi=kitapLi.find("div",{"class":"p13n-sc-truncate p13n-sc-line-clamp-1"})
        kitapPuani=kitapLi.find("span",{"class":"a-icon-alt"})
        kitapFiyat=kitapLi.find("span",{"class":"p13n-sc-price"})
        print("Kitap Ismi:",kitapIsmi.text.strip("\n \t"))
        print("Kitap Puani:",kitapPuani.text.strip("\n \t"))
        print("Kitap Fiyati:",kitapFiyat.text.strip("\n \t"))
        print("**********************************************")
else:
    print("Hata meydana geldi!")
```

### Requests Modülüne Neden İhtiyacımız Var?

Requests modülünü kullanarak web üzerindeki HTTP isteklerimizi(GET,POST,PUT,DELETE) yönetebileceğiz.

```python
import requests
```

#### GET Metodu ile Sayfa İçeriğini Alma ve Parametre Gönderme

GET Metodu ile parametre gönderdiğimiz zaman, gönderdiğimiz parametreler URL yapısı içerisinde bulunuyor.

```python
sayfaURL="https://httpbin.org/get"
sayfaCevabi=requests.get(sayfaURL)
sayfaCevabi.status_code
```

```python
sayfaCevabi.url
```

```python
sayfaCevabi.content
```

```python
sayfaCevabi=requests.get(sayfaURL,params={"kullanici_adi":"ogunbrnc","sifre":"123456"})
sayfaCevabi.status_code
```

```python
sayfaCevabi.url
```

```python
sayfaCevabi.content
```

#### Post Metodu ile Sayfa İçeriğini Alma ve Sayfaya Veri Gönderme

```python
sayfaCevabi.url
```

```python
sayfaCevabi.content
```

### BeautifulSoup Modülüne Neden İhtiyacımız Var?

HTML içeriğini çektiğimiz web siteyi parçalamak için kullanıyoruz

```python
from bs4 import BeautifulSoup
```

#### BeautifulSoup Obje Oluşturarak Parçalama İşlemi ve Prettify Fonksiyonu

HTML içeriğini aldığımız web siteyi BeautifulSoup ile parçalayabiliyoruz ve prettify fonksiyonunu kullanarak daha düzgün görünmesini sağlayabiliyoruz.

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    print(soup.prettify())
    
```

#### Belirli Bir Etiket Değerini Alma ve find Fonksiyonu

find fonksiyonu ile birlikte oluşturduğumuz BeautifulSoup obje üzerinde etiket araması yapabiliriz, eşleşen tek bir etiket bize verilir.

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    print(soup.find("h1"))
```

#### Belirli Bir Etiket Değerlerini Alma ve find_all Fonksiyonu

find_all fonksiyonu ile birlikte oluşturduğumuz BeautifulSoup obje üzerinde etiket araması yapabiliriz, eşleşen etiketler liste halinde bize verilir.

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for icerik in soup.find_all("a"):
        print(icerik)
```

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for icerik in soup.find_all("span",{"class":"img"}):
        print(icerik)
```

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for icerik in soup.find_all("a",{"class":"active"}):
        print(icerik)
```

#### Etiket İçerisinden Belirli Değerleri Alma ve get Fonksiyonu

get fonksiyonu ile birlikte find_all fonksiyonu ile elde ettiğimiz liste üzerinden belirli kısımları elde edebiliriz.

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for icerik in soup.find_all("a"):
        print(icerik)
```

```python
sayfaURL="https://www.sondakika.com/"
sayfaCevabi=requests.get(sayfaURL)
if sayfaCevabi.status_code==200:
    htmlIcerik=sayfaCevabi.content
    soup=BeautifulSoup(htmlIcerik,"html.parser")
    for icerik in soup.find_all("a"):
        print(icerik.get("title"))
```

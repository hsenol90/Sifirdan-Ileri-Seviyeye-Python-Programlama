## Request Modülü ile HTTP İşlemleri


### HTTP Nedir?

HTTP(Hyper Text Transfer Protocol), verilerin sunucudan kullanıcıya nasıl aktarılacağını kontrol eden protokoldür. İstek ve yanıt şeklinde çalışır, sunucuya bir istek gönderilir ve o sunucudan olumlu veya olumsuz bir cevap döner.

```python
import urllib.request
```

### urlopen Fonksiyonu ile Siteye Bağlanma

```python
siteURL="https://httpbin.org/"
```

```python
html=urllib.request.urlopen(siteURL)
html
```

### read Fonksiyonu ile Sitenin HTML Bilgisini Alma

```python
html.read()
```

### Header Bilgisi Gönderme

Bazı sitelerde bu kadar kolay şekilde bağlanıp bilgi çekemiyoruz, iznimiz olmuyor. Bunun için de User Agent içerisinde tanımlı olan işletim sistemi, tarayıcı ve cihaz bilgilerini header bilgisiyle birlikte siteye göndermemiz gerekebiliyor.

```python
siteURL="https://eksisozluk.com/"
html=urllib.request.urlopen(siteURL)
html
```

```python
siteURL="https://eksisozluk.com/"
header_bilgileri={
    "User-Agent":"Mozilla/5.0 (Linux; Android 10; SM-A205U) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.62 Mobile Safari/537.36."
}
istek=urllib.request.Request(siteURL,headers=header_bilgileri)
html=urllib.request.urlopen(istek)
html.read()
```

### Sitelere Veri Gönderme İşlemi

```python
veriler={
    "kullanici_adi":"ogunbrnc",
    "sifre":"123456",
    "email":"ogunbrnc@gmail.com",
    "telefon":"05521234567",
    "bolum":"Bilgisayar Muhendisligi"
}
```

#### urlencode Fonksiyonu ve Şifrelenmiş URL Oluşturma

```python
import urllib.parse
sifreliVeriler=urllib.parse.urlencode(veriler)
sifreliVeriler
```

Veriler arasına & koyulmuş ve  @ karakteri ise %40 ile, boşluk karakteri ise + ile değiştirilmiş

#### Get Metodu ile Veri Gönderme

Get metodu ile gönderilen veriler URL yapısı ile birleştirilir, yani URL üzerinde saklanır ve URL üzerinde gözükür. Sadece yazı formatında veriler gönderebiliriz. Sitenin URL yapısı ve gönderdiğimiz veriler birbirinden "?" işareti ile ayrılır.

```python
siteURL="http://httpbin.org/get"
veriler={
    "kullanici_adi":"ogunbrnc",
    "sifre":"123456",
    "email":"ogunbrnc@gmail.com",
    "telefon":"05521234567",
    "bolum":"Bilgisayar Muhendisligi"
}
sifreliVeriler=urllib.parse.urlencode(veriler)
sonSiteURL=siteURL+"?"+sifreliVeriler
print(sonSiteURL)
html=urllib.request.urlopen(sonSiteURL)
print(html.read())
```

#### Post Metodu ile Veri Gönderme

Post metodu ile gönderilen veriler URL yapısı ile birleştirilmez, yani URL üzerinde saklanmaz ve URL üzerinde gözükmez.Sadece yazı formatında veriler değil, resim,ses,dosya gibi farklı türde veriyi POST metodu ile gönderebiliriz. 

```python
siteURL="http://httpbin.org/post"
veriler={
    "kullanici_adi":"ogunbrnc",
    "sifre":"123456",
    "email":"ogunbrnc@gmail.com",
    "telefon":"05521234567",
    "bolum":"Bilgisayar Muhendisligi"
}
sifreliVeriler=urllib.parse.urlencode(veriler)
sifreliVeriler=sifreliVeriler.encode("utf-8")

istek=urllib.request.Request(siteURL,data=sifreliVeriler)
html=urllib.request.urlopen(istek)

html.read()
```

### HTML Etikete Göre İçerik Okuma

```python
siteURL="https://www.google.com/"
html=urllib.request.urlopen(siteURL)
icerik=html.read()
icerik
```

```python
import re

siteURL="https://www.sondakika.com/"
html=urllib.request.urlopen(siteURL)
icerik=str(html.read())
etiketTanimi="<title>(.*?)</title>"
aramaSonucu=re.search(etiketTanimi,icerik)

if aramaSonucu:
    etiket=aramaSonucu.group(0)
    icerik=aramaSonucu.group(1)
    
    print("Etiket:"+etiket)
    print("Icerik:"+icerik)

```
